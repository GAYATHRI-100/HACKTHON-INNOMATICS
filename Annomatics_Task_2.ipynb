Problem 2: Highest Salary from Employee Data


```python
employees = {}  #At initial create a new dictionary

n = int(input("Enter number of employees: "))

for i in range(n):
    name = input("Enter employee name: ")
    salary = float(input("Enter salary: "))
    employees[name] = salary

# Find highest salary
highest_paid_employee = max(employees, key=employees.get)
highest_salary = employees[highest_paid_employee]

print("Highest Salary:", highest_paid_employee, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values


```python
# Given list of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with the first element
max_number = numbers[0]
min_number = numbers[0]

# Loop through the list
for num in numbers:
    if num > max_number:
        max_number = num
    if num < min_number:
        min_number = num

# Print results
print("List:", numbers)
print("Max:", max_number)
print("Min:", min_number)
```

Problem 4: Count Products Above a Price Threshold

```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Initialize count at initial is to 0
count = 0

# Loop through prices
for i in prices:

  if i > 1000:

    count = count + 1

# Print result
print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage


```python
# Attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count total days and present days
total_days = len(attendance)
present_count = 0

for i in attendance:

  if i == "P":

    present_count = present_count + 1

# Calculate attendance percentage
attendance_percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", attendance_percentage)
```

Problem 7: Count Character Frequency


```python
# Input string
text = input()

# Create an empty dictionary
char_count = {}

# Loop through each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

# Print the result
print(char_count)

```

Problem 10: Calculate Average Marks


```python
# List of marks
marks = [70, 75, 80, 65]

# Calculate sum and total subjects
total = sum(marks)
no_subjects = len(marks)

# Calculate average
average_marks = total / no_subjects

# Print result
print("Average Marks:", average_marks)
```
