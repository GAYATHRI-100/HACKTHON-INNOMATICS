problem statement 1:Employees performance Bonus Eligibility

```python
# Input dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
maximum_score = max(employees.values())

# Step 2: Find all employees with the highest score
top_performers = [name for name, score in employees.items() if score == maximum_score]

# Step 3: result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {maximum_score})")
```

Problem statement 2: Search query keyword Analysis

```python

query = "Buy mobile phone buy phone online"

# Step 1: Convert the entire sentence to lowercase
query = query.lower()

# Step 2: Remove punctuation marks (.,!? etc.)
import string
for symbol in string.punctuation:
    query = query.replace(symbol, "")

# Step 3: Split the sentence into individual words
words = query.split()

# Step 4: Count how many times each word appears
keyword_count = {}

for word in words:
    # If the word is already in the dictionary, increase its count
    if word in keyword_count:
        keyword_count[word] += 1
    else:
        # Otherwise, start the count at 1
        keyword_count[word] = 1

# Step 5: Keep only keywords searched more than once
repeated_keywords = {}

for word, count in keyword_count.items():
    if count > 1:
        repeated_keywords[word] = count

# Step 6: Print the result
print(repeated_keywords)
```

Problem statement 3: Sensor Data Validation

```python
# Sensor readings list (index = hour, value = sensor reading)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid sensor readings (hour, value)
valid_readings = []

# Loop through the list with index (hour) and value
for hour, value in enumerate(sensor_readings):
    # Check if the reading is even (valid)
    if value % 2 == 0:
        valid_readings.append((hour, value))

# Display the result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4:Email Domain Usage Analysis


```python
# List of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Dictionary to store domain counts
domain_count = {}

# Step 2: Extract domain names and count them
for email in emails:
    # Split email into name and domain
    domain = email.split("@")[1]

    # Count each domain
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 3: Calculate total number of emails
total_emails = len(emails)

# Step 4: Calculate and display percentage usage

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

Problem Statement 5: Sales Spike Detection


```python
# Daily sales data for each day
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Find the average daily sales
total_sales = sum(sales)
number_of_days = len(sales)
average_sales = total_sales / number_of_days

# Step 2: Define spike limit (30% above average)
spike_limit = average_sales * 1.29

# Step 3: Check each day for sales spikes

for day, sale in enumerate(sales, start=1):   # start=1 to show Day 1, Day 2, ...

    # If sale is greater than 30% above average, it's a spike
    if sale > spike_limit:
        print(f"Day {day}: {sale}")



```

Problem Statement 6: Duplicate User ID Detection


```python
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Dictionary to store user ID counts
user_count = {}

# Step 2: Count each user ID
for user in user_ids:
    if user in user_count:
        user_count[user] += 1
    else:
        user_count[user] = 1

# Step 3: Print only duplicate user IDs
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
