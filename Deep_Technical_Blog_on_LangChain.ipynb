```python
from langchain.memory import ConversationBufferMemory
from langchain.prompts import PromptTemplate

memory = ConversationBufferMemory(return_messages=False)

prompt = PromptTemplate.from_template("""
You are a helpful AI tutor.

Conversation so far:
{history}

User: {input}
AI:
""")

chain = prompt | llm

print("\n🔹 Memory Output:")

# ---- TURN 1 ----
user_input = "Hello"
history = memory.load_memory_variables({})["history"]

response = chain.invoke({
    "history": history,
    "input": user_input
})

memory.save_context(
    {"input": user_input},
    {"output": response}
)

print(response)


# ---- TURN 2 ----
user_input = "What am I learning?"
history = memory.load_memory_variables({})["history"]

response = chain.invoke({
    "history": history,
    "input": user_input
})

memory.save_context(
    {"input": user_input},
    {"output": response}
)

print(response)
```

```python
import re

# ---- SAFE CALCULATOR ----
def calculator(expr):
    try:
        return eval(expr)
    except:
        return None

# ---- FIXED EXTRACTION ----
def extract_expression(text):
    # extract everything after "What is"
    match = re.search(r'what is\s*(.*)', text.lower())

    if match:
        expr = match.group(1)

        # clean unwanted characters
        expr = re.sub(r'[^0-9\+\-\*/\(\)\.\s]', '', expr)

        return expr.strip()

    return None

# ---- DETECTOR ----
def is_math_expression(text):
    return "what is" in text.lower()

# ---- AGENT ----
def agent(query):
    print("User Query:", query)

    if is_math_expression(query):
        expr = extract_expression(query)

        print("🔧 Tool Used: Calculator")
        print("Expression:", expr)

        result = calculator(expr)

        return f"The result of {expr} is {result}."

    return "No computation required."

# ---- RUN ----
response = agent("What is (45 * 2) + 10?")
print("\n🔹 FINAL OUTPUT:\n")
print(response)
```
