Task 3: Stress Testing


```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Lowercase
    text = text.lower()

    # Strong normalization: reduce repeated chars to ONE
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # Remove special characters
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    # Remove short words except "no", "not"
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

```python
text = "I absolutely looooved this product 😍😍"
tokens, sentence = preprocess_text(text)

print("Tokens:", tokens)
print("Cleaned Sentence:", sentence)
```

Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)

        # Collect all tokens
        all_tokens.extend(tokens)

        # Collect cleaned sentences
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(test_sentences)

print("All Tokens:")
print(output["tokens"])

print("\nCleaned Sentences:")
for sent in output["clean_sentences"]:
    print(sent)
```
