2.Pass / Fail Analyzer


```python
marks = [45, 78, 90, 33, 60] # student marks details
pass_count = 0  #initialize count values
fail_count = 0
for i in marks:
  if i >= 50:
    pass_count += 1
  else:
    fail_count += 1
print("Total number of students marks: ",len(marks))
print("Total number of students passed: ", pass_count)
print("Total number of students failed: ", fail_count)

```

3.Simple Data Cleaner


```python

names = [" Alice ", "bob", " CHARLIE "]


updated_nameslist = []

for name in names:
    name = name.strip()      # Remove extra spaces
    name = name.lower()      # Convert to lowercase
    updated_nameslist.append(name)


print("Names List:", updated_nameslist)
```

4.Message Length Analyzer


```python

messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for msg in messages:
    length = len(msg)   # Calculate message length

    print(f"Message: '{msg}'")
    print(f"Length: '{length}'")

    # Flag long messages
    if length > 10:
        print("Long message detected\n")
    else:
        print("Normal message\n")
```

5. Error Message Detector


```python

logs = ["INFO", "ERROR", "WARNING", "ERROR"]


error_count = 0

# Scan logs to detect errors
for i in logs:
    if i == "ERROR":
        error_count += 1

# Print final result
print("Total ERROR messages detected:", error_count)
```
