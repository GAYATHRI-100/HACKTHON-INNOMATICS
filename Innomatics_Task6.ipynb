Problem 1: Smart Parking Lot Management System

```python
def smart_parking_lot_management_system(capacity,logs):
  parked =0
  parked_usage =0
  for log in logs:
    if log =="IN":
      parked +=1
    elif log == "OUT" and parked > 0:

      parked-=1
    if parked > parked_usage:
      parked_usage = parked
  print("currently parked vehical:",parked)
  print("maximum parked vehical:",parked_usage)
  if parked > capacity:
    print("Parking Status:Capacity Exceeded")
  else:
    print("Parking Status:Available")
capacity = int(input("Enter the capacity :"))
n =  int(input("Enter the number of logs :"))
logs = []
for i in range(n):
  log = input(f"Enter the logs{i+1}(IN/OUT) :")
  logs.append(log)
smart_parking_lot_management_system(capacity,logs)




```

 Problem Statement 2:Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance,traffic,weather):
  base_speed = 0.4
  time = distance / base_speed
  if  traffic == "low":
    traffic_delay = 0
  elif traffic == "normal":
    traffic_delay = 10
  elif traffic == "high":
    traffic_delay = 20
  else:
    traffic_delay = 0
  if weather == "clear":
    weather_delay = 0
  elif weather == "rainy":
    weather_delay = 15
  elif weather == "stromy":
    weather_delay = 30
  else:
    weather_delay = 0
  final_time = time + traffic_delay + weather_delay
  print("Estimated delivery time:",int(final_time), "minutes")
distance = int(input("Enter the distance(km):"))
traffic = input("Enter the traffic(low/normal/high):")
weather = input("Enter the weather(clear/rainy/stromy):")
estimate_delivery_time(distance,traffic,weather)

```

Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def seat_booking_analysis(total_seats, booked_list):
  booked_seats = sum(booked_list)
  occupancy = (booked_seats / total_seats) * 100


  if occupancy == 100:
    status = "Housefull"
  elif occupancy >= 75:
    status = "Almost Full"
  elif occupancy >= 40:
    status = "Moderate Booking"
  else:
    status = "Seats Available"


  if occupancy >= 90:
    suggestion = "Suggest opening additional shows"
  else:
    suggestion = "No extra show needed"
  print("Occupancy:", int(occupancy), "%")
  print("Show Status:", status)
  print("Suggestion:", suggestion)

total_seats = int(input("Enter Total Seats: "))

n = int(input("Enter number of booked entries: "))
booked_list = []

for i in range(n):
    booked_list.append(int(input("Enter booking (1 for booked): ")))


seat_booking_analysis(total_seats, booked_list)


```

Problem Statement 4:Cloud Server Load Classification System


```python
def cloud_server_classify(cpu_readings):
  total = 0
  for cpu in cpu_readings:
    total = total + cpu
  average = total/(len(cpu_readings))
  if average < 50:
    status = "Normal"
  elif 50 <= average <= 80:
    status = "Warning"
  else:
    status = "Critical"
  print("Average CPU Usage:", int(average), "%")
  print("Server Status:", status)
n=int(input("Enter the number of cpu readings:"))
cpu_readings = []
for i in range(n):
  value = int(input(f"Enter the cpu reading{i+1}:"))
  cpu_readings.append(value)
cloud_server_classify(cpu_readings)

```

Problem Statement 5:Smart Classroom Resource Usage Monitor
Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.


```python
def resource_usage(resources):
  limit = 7
  overused = []
  for resource,hours in resources.items():
    if hours > limit:
      overused.append(resource)
  if overused:
      print("overused resources:",",".join(overused))
      print("Energy Alert:" ,"yes")
  else:
      print("overused resources:","None")
      print( "Energy Alert:","No")
n=int(input("Enter the number of resources:"))
resources = {}
for i in range(n):
  name = input(f"Enter the name of resource{i+1}:")
  value = int(input(f"Enter the hours for {name}:"))
  resources[name] = value
resource_usage(resources)
```

 Problem Statement 6: Online Event Registration Capacity Controller

```python

def registration_capacity(capacity,registrations):
  confirmed=0
  waitedlist =0
  for i in range(registrations):
    if confirmed < capacity:
      confirmed += 1
    else:
      waitedlist +=1
  if confirmed >= capacity:
    status = "closed"
  else:
    status="open"
  print("confirmed registration:",confirmed)
  print("waitedlist:",waitedlist)
  print("registation status:",status)
capacity = int(input("Enter the capacity:"))
registrations = int(input("Enter the number of registrations:"))
registration_capacity(capacity,registrations)
```
