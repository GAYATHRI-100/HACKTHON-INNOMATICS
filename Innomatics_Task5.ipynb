Problem 1: Social Media – Post Engagement Analyzer


```python
def analyze_engagement(likes_post):
  total_likes = 0
  for likes in likes_post:
    total_likes += likes
  if total_likes >= 1000:
    status = "viral post"
  else:
    status = "normal post"
  print("Total likes:", total_likes)
  print("Status:", status)
likes_data = [250,100,100,300,300]
analyze_engagement(likes_data)
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_checker(rainfall_list,required_list):
  total = 0
  for rain in rainfall_list:
    total += rain
  average = total / len(rainfall_list)
  print(f"Average rainfall: {average}")
  if average >= required_list:
    status = "adequate"
  else:
    status = "inadequate"
  print("Rainfall Status:", status)
n = int(input("Enter the number of days: "))
rainfall_list = []
for i in range(n):
  value = float(input(f"Enter the rainfall for day {i+1}: "))
  rainfall_list.append(value)
required_list = float(input("Enter the required rainfall: "))
rainfall_checker(rainfall_list,required_list)

```

Problem 6: Agriculture – Premium Crop Price Filter


```python
def  premium_crop_filter(prices):
  premium_thresold = 2000
  premium_crops = [price for price in prices if price > premium_thresold]
  return premium_crops
n=int(input("Enter the number of crops:"))
prices=[]
for i in range(n):
  price = int(input(f"Enter the price of crop{i+1}:"))
  prices.append(price)
premium_crops=premium_crop_filter(prices)
print("premium crops:",premium_crops)

```

Problem 8: Banking – Daily Transaction Limit Checker

```python
# Function to validate transaction amount
def validate_transaction(amount):
    daily_limit = 50000  # Daily transaction limit

    if amount <= daily_limit:
        return "Approved"
    else:
        return "Rejected"

# User input
amount = int(input("Enter transaction amount: "))

# Function call
status = validate_transaction(amount)

# Output
print("Transaction Amount:", amount)
print("Transaction Status:", status)
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
# Function to calculate attendance percentage and eligibility
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Loop through attendance list
    for day in attendance_list:
        if day == 1:   # 1 means Present, 0 means Absent
            present_days += 1

    # Calculate percentage
    percentage = (present_days / total_days) * 100

    # Check eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return percentage, status


# User input
n = int(input("Enter total number of days: "))
attendance = []

print("Enter attendance (1 = Present, 0 = Absent):")
for i in range(n):
    value = int(input(f"Day {i+1}: "))
    attendance.append(value)

# Function call
percentage, status = check_attendance(attendance)


print("Attendance Percentage:", percentage)
print("Exam Eligibility:", status)
```

```python
# Function to calculate electricity bill and usage classification
def calculate_electricity_bill(units):
    bill = 0

    # Billing slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# User input
units = int(input("Enter total electricity units consumed: "))

# Function call
total_bill, usage_status = calculate_electricity_bill(units)

# Output
print("Total Units Consumed:", units)
print("Electricity Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
