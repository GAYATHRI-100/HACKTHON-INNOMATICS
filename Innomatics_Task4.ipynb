Problem 1: Mobile Recharge Validation System


```python
# Function to validate recharge amount
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print(" Minimum recharge should be ₹50")
        return False

    if amount not in valid_plans:
        print(" Invalid plan! Available plans are:", valid_plans)
        return False

    print(" Recharge successful! Plan activated.")
    return True



while True:
    try:
        recharge_amount = int(input("Enter recharge amount (₹): "))

        # Call validation function
        if validate_recharge(recharge_amount):
            break   # Exit loop if recharge is valid

        # Ask user if they want to retry
        retry = input("Do you want to try again? (yes/no): ").lower()
        if retry != "yes":
            print("Recharge cancelled. Thank you!")
            break

    except ValueError:
        print(" Please enter a valid numeric amount.")
```

Problem 2: Inventory Reorder Alert System


```python
# Function to check stock and print alerts
def check_inventory(stock_data):
    print(" Inventory Status Report\n")

    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f" Reorder Alert! {product} stock is low: {quantity} units")
        else:
            print(f" Stock OK for {product}: {quantity} units")


# Dictionary of products and stock quantities
inventory = {
    "Rice Bags": 10,
    "Sugar Packs": 25,
    "Oil Bottles": 12,
    "Soap Bars": 50,
    "Wheat Flour": 8
}

# Call the function
check_inventory(inventory)
```

 Problem 3: Student Result Processing System


```python
# Function to process student marks
def student_result(marks):
    total = 0

    # Calculate total marks using loop
    for m in marks:
        total += m

    # Calculate average
    average = total / len(marks)


    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result


# List of student marks
student_marks = [60, 45, 70, 55, 40]

# Call the function
avg, status = student_result(student_marks)


print("Student Marks:", student_marks)
print("Average Marks:", avg)
print("Result:", status)
```

Problem 4: Cab Fare Estimator with Retry Option


```python
# Function to calculate cab fare
def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km_rate = 12


    fare = base_fare + (distance * per_km_rate)


    if peak_hour == "yes":
        fare = fare + (fare * 0.25)

    return fare



while True:
    try:
        km = float(input("Enter distance traveled (in km): "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        total_fare = calculate_fare(km, peak)

        print(f" Total Cab Fare = ₹{total_fare:.2f}")


        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using the Cab Fare Estimator!")
            break

    except ValueError:
        print(" Please enter valid numeric distance.")
```

Problem 5: Employee Attendance Eligibility Checker


```python
# Function to check attendance eligibility
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / total_days) * 100

    # Check eligibility
    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"

    return percentage, result


# Attendance list (P = Present, A = Absent)
attendance = ["P", "A", "P", "P", "P", "A", "P", "P"]


percent, status = check_attendance(attendance)


print("Attendance Record:", attendance)
print("Attendance Percentage:", percent, "%")
print("Eligibility Status:", status)
```

Problem 6: Password Strength Checker


```python
# Function to check password strength
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    # Check each character using loop
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # Check all rules
    if len(password) < 8:
        return "Weak Password: Minimum length should be 8"

    if not has_digit:
        return "Weak Password: Must contain at least one digit"

    if not has_special:
        return "Weak Password: Must contain at least one special character (@, #, $)"

    return "Strong Password"



pwd = input("Enter your password: ")

result = check_password(pwd)
print("Password Status:", result)
```
