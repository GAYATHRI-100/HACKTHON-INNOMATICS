Problem 1: Smart Light Controller

```python
class Light:
    def __init__(self, name):
        self.name = name
        self.is_on = False

    def on(self):
        self.is_on = True
        print(f"{self.name} is ON")

    def off(self):
        self.is_on = False
        print(f"{self.name} is OFF")


Light("Bedroom Light").on()
```

Problem 2: Employee ID Card System


```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def show_card(self):
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


rahul = Employee("Rahul", "EMP102", "IT")
rahul.show_card()
```

Problem 3: Mobile Contact Record


```python
class MyContact:
    def __init__(self, name, number):
        self.name = name
        self.number = number

    def show_contact(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.number}")


anita = MyContact("Anita", "9876543210")
anita.show_contact()
```

Problem 4: Product Price Tag Generator


```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
item = Product("Headphones", "2499")
item.tag()
```

Problem 5: Movie Rating Display System


```python
class Film:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")
Film("Inception", "4.8").display()
```

Problem 6: Delivery Address Manager


```python
class Package:
    def __init__(self, name, location):
        self.name = name
        self.location = location

    def label(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.location}")
Package("Suman", "Hyderabad").label()
```
